```python
from pandas_datareader import data as pdr

import yfinance as yf
import requests
from bs4 import BeautifulSoup
# yf.pdr_override() # <== that's all it takes :-)

# download dataframe
def get_stock_data(stock_name):
    # data = pdr.get_data_yahoo(stock_name, start="2007-01-01", end="2024-01-01")
    # data = data.reset_index()
    # data.Date = pd.to_datetime(data.Date)
    # # data.set_index('Date', inplace = True)
    data = yf.Ticker(stock_name).history(period='1y')
    data = data.reset_index()
    data.Date = pd.to_datetime(data.Date)
    return data

def get_sp500_stocks_tickers_from_wiki():
    url = 'https://en.wikipedia.org/wiki/List_of_S%26P_500_companies'
    response = requests.get(url
                            , headers={'User-Agent': 'Mozilla/5.0'})
    soup = BeautifulSoup(response.text, 'html.parser')
    table = soup.find_all('table')[0]
    tickers = []

    for row in table.find_all('tr')[1:]:
        tickers.append(row.find_all('td')[0].text.strip())
    return tickers
```

```python
if __name__ == "__main__":

    tickers = get_sp500_stocks_tickers_from_wiki()

    dfs = []

    for ticker in tqdm(tickers):
        try:
            df = get_stock_data(ticker)
            df['stock_name'] = ticker
            dfs.append(df)
        except:
            print('error', ticker)

        # dfs = pd.concat([dfs,df], axis=0)

    final_df = pd.concat(dfs, axis=0)
    final_df

```

```python
final_df['Date'] = final_df['Date'].apply(lambda x: x.strftime('%Y-%m-%d'))
final_df = final_df.drop('Adj Close', axis=1)
final_df = final_df.round(5)
final_df.head()
```
